```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.flipkart.com/samsung-mobile-store?otracker=nmenu_sub_Electronics_0_Samsung"
url
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.0.0"}
headers
```

```python
page = requests.get(url, headers=headers)
page          
```

```python
soup1 = BeautifulSoup(page.content, "html.parser")
soup1
```

```python
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
soup2

```

```python
models = soup2.find("div",class_="_37K3-p").find_all("div",class_="_3YgSsQ")
models

```

```python
import csv 

header = ['Model', 'Price', 'Rating']

with open('flipkart2.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
```

```python
for each in models:
    model = each.find("a",class_="s1Q9rs").text.strip()
    price = each.find("div",class_="_30jeq3").text.strip()[1:]
    rating = each.find("div",class_="_3LWZlK").text.strip()
    with open('flipkart2.csv', 'a+', newline='', encoding='UTF8') as f:
        data = [model,price,rating]
        writer = csv.writer(f)
        writer.writerow(data)

    
```
